```python
import pandas as pd
import jsonlines
```

```python
pwd_based_paths = "../build/volumes/data/search_results/password_based_authentication_fastapi_search_results.jsonl"
verifiable_token_paths = "../build/volumes/data/search_results/verifiable_token_authentication_fastapi_search_results.jsonl"

def calculate_success_counts(context_paths):
    with jsonlines.open(context_paths) as reader:
        data = [obj for obj in reader]
        
    dataframe = pd.DataFrame(data)
    # Filter for successful entries
    successful = dataframe[dataframe['success'] == True]
    # Group by role and count unique repositories
    return successful.groupby('role')['repo'].nunique().sort_values(ascending=False)
```

```python
calculate_success_counts(pwd_based_paths)
```

```python
calculate_success_counts(verifiable_token_paths)
```
